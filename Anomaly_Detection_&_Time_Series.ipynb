#Q1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:

Anomaly Detection is the process of identifying unusual or abnormal data points that differ significantly from the normal behavior of a dataset. These unusual observations are called anomalies or outliers. Anomaly detection is widely used in fraud detection, network security, healthcare, manufacturing, and fault detection.

Types of Anomalies
1. Point Anomaly

A point anomaly is a single data point that is significantly different from the rest of the data.

Example:
A customer's normal daily bank transaction is ₹5,000, but one day a transaction of ₹5,00,000 occurs.

2. Contextual Anomaly

A contextual anomaly is abnormal only in a specific context, such as time or location.

Example:
A temperature of 30°C is normal in summer but unusual during winter.

3. Collective Anomaly

A collective anomaly occurs when a group of related data points behaves abnormally, even though individual points may appear normal.

Example:
Several failed login attempts within a few minutes may indicate a cyberattack.

Applications:

Fraud Detection
Network Intrusion Detection
Medical Diagnosis
Manufacturing Quality Control
Predictive Maintenance
#Q2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.

Answer:

Algorithm	Approach	Suitable Use Cases
Isolation Forest	Randomly isolates anomalies using decision trees.	Large datasets, fraud detection, anomaly detection.
DBSCAN	Groups dense regions and labels sparse points as outliers.	Spatial data, GPS data, customer segmentation.
Local Outlier Factor (LOF)	Measures the local density around each data point to identify outliers.	Datasets with varying densities, sensor data, medical data.
Comparison
Isolation Forest
Fast and efficient for large datasets.
Works well with high-dimensional data.
Does not require labeled data.
DBSCAN
Detects clusters of arbitrary shapes.
Identifies noise automatically.
Sensitive to parameter selection.
LOF
Detects local anomalies effectively.
Suitable when data density varies.
Computationally expensive for very large datasets.
#Q3. What are the key components of a Time Series? Explain each with one example.

Answer:

A Time Series is a sequence of observations recorded over time. It consists of the following components:

1. Trend

Trend represents the long-term increase or decrease in data.

Example:
The number of online shoppers increases every year.

2. Seasonality

Seasonality refers to regular repeating patterns over fixed intervals.

Example:
Ice cream sales increase every summer.

3. Cyclical Component

Cyclical variation occurs over long periods due to economic or business cycles.

Example:
House prices rise and fall during economic expansions and recessions.

4. Irregular (Random) Component

Random variations are unpredictable fluctuations caused by unexpected events.

Example:
A sudden drop in airline passengers due to a natural disaster.

#Q4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:

A stationary time series is one whose statistical properties, such as mean, variance, and autocorrelation, remain constant over time. Many forecasting models like ARIMA require stationary data.

Testing Stationarity

The most commonly used test is:

Augmented Dickey-Fuller (ADF) Test
p-value < 0.05: Series is stationary.
p-value > 0.05: Series is non-stationary.
Methods to Make a Series Stationary
Differencing
Log Transformation
Moving Average Smoothing
Seasonal Differencing
Removing Trend and Seasonality

Example:
Monthly airline passenger data can be transformed into a stationary series using first-order differencing.

#Q5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

Answer:

Model	Structure	Application
AR (AutoRegressive)	Uses previous observations to predict future values.	Forecasting stationary data.
MA (Moving Average)	Uses previous forecast errors.	Modeling random shocks in stationary data.
ARIMA	Combines AR + I (Differencing) + MA.	Forecasting non-stationary time series after differencing.
SARIMA	ARIMA + Seasonal components.	Seasonal data such as monthly sales or airline passengers.
SARIMAX	SARIMA + External (exogenous) variables.	Forecasting using seasonal data along with external factors like weather, holidays, or promotions.
Summary
AR: Depends on past values.
MA: Depends on past errors.
ARIMA: Suitable for non-seasonal time series after differencing.
SARIMA: Best for seasonal time series.
SARIMAX: Best when external variables influence the time series.

#Q6. Load a time series dataset (AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.

```python
# Q6: Time Series Decomposition

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Create date index
data['time'] = pd.date_range(start='1949-01', periods=len(data), freq='MS')
data.set_index('time', inplace=True)

# Rename column
series = data['value']

# Plot original series
plt.figure(figsize=(10,5))
plt.plot(series)
plt.title("AirPassengers Dataset")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()

# Decompose time series
result = seasonal_decompose(series, model='multiplicative')

# Plot decomposition
result.plot()
plt.show()
```

#Q7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
# Q7: Isolation Forest

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import IsolationForest

# Load dataset
data = load_breast_cancer()

X = data.data[:, :2]   # Use first two features

# Train Isolation Forest
model = IsolationForest(
    contamination=0.05,
    random_state=42
)

pred = model.fit_predict(X)

# Plot
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=pred,
    cmap='coolwarm'
)

plt.title("Isolation Forest Anomaly Detection")
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])

plt.show()
```

#Q10. Real-Time Energy Demand Forecasting and Anomaly Detection
Explanation

Suppose you are working as a Data Scientist for a power grid monitoring company. The system receives electricity consumption data every 15 minutes along with information such as timestamp, region, weather conditions, and energy usage.

Workflow
1. Anomaly Detection
Use Isolation Forest for large streaming datasets because it is fast and scalable.
Use LOF when detecting local anomalies in regions with different consumption patterns.
Use DBSCAN to identify unusual clusters or abnormal consumption behavior.
2. Time Series Forecasting
ARIMA: Suitable for non-seasonal data.
SARIMA: Best for seasonal energy demand.
SARIMAX: Best choice because weather conditions, holidays, and region affect electricity usage.
3. Validation and Monitoring
Split data into training and testing sets.
Evaluate forecasting using:
MAE (Mean Absolute Error)
RMSE (Root Mean Squared Error)
MAPE (Mean Absolute Percentage Error)
Continuously monitor prediction errors.
Retrain the model periodically using newly collected data.
4. Business Benefits
Detect equipment failures early.
Prevent power outages.
Optimize energy generation and distribution.
Reduce operational costs.
Improve customer satisfaction.
Support better planning and resource allocation.
Python Code

```python
# Q10: Energy Demand Forecasting & Anomaly Detection

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Generate synthetic energy usage data
np.random.seed(42)

time = pd.date_range(
    start='2024-01-01',
    periods=200,
    freq='15min'
)

energy = np.random.normal(100, 10, 200)

# Add anomalies
energy[30] = 180
energy[120] = 40
energy[170] = 170

df = pd.DataFrame({
    "Timestamp": time,
    "Energy": energy
})

# ---------------------------
# Isolation Forest
# ---------------------------
model = IsolationForest(
    contamination=0.02,
    random_state=42
)

df["Anomaly"] = model.fit_predict(df[["Energy"]])

# Plot anomalies
plt.figure(figsize=(12,5))

plt.plot(df["Timestamp"], df["Energy"], label="Energy Usage")

plt.scatter(
    df[df["Anomaly"]==-1]["Timestamp"],
    df[df["Anomaly"]==-1]["Energy"],
    color="red",
    label="Anomalies"
)

plt.legend()
plt.title("Energy Usage with Detected Anomalies")
plt.show()

# ---------------------------
# SARIMA Forecast
# ---------------------------
sarima = SARIMAX(
    df["Energy"],
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

result = sarima.fit()

forecast = result.forecast(20)

plt.figure(figsize=(12,5))

plt.plot(df["Energy"], label="Original")
plt.plot(
    range(len(df), len(df)+20),
    forecast,
    color="green",
    label="Forecast"
)

plt.legend()
plt.title("SARIMA Forecast")
plt.show()
```
